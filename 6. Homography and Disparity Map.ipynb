```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
source_pts = np.array([[121,111],[460,140],[470,610],[44,580]])
destination_pts = np.array([[270,450],[235,789],[612,534],[99,999]])

homography_matrix, status = cv2.findHomography(source_pts, destination_pts)
print(homography_matrix)
```

<font size=3 color='darkblue'>Disparity Map

```python
left = cv2.imread("DIsparity_images/left.jpeg")
right = cv2.imread("DIsparity_images/right.jpeg")

left_gray = cv2.cvtColor(left, cv2.COLOR_BGR2GRAY)
right_gray = cv2.cvtColor(right, cv2.COLOR_BGR2GRAY)

fig, ax = plt.subplots(1,2,figsize=(7,7))

ax[0].imshow(cv2.cvtColor(left_gray,cv2.COLOR_BGR2RGB))
ax[0].set_title("Left Frame")

ax[1].imshow(cv2.cvtColor(right_gray,cv2.COLOR_BGR2RGB))
ax[1].set_title("Right Frame")

plt.tight_layout()
plt.show()
```

```python
stereo = cv2.StereoBM_create(numDisparities=32, blockSize=25)
disparity = stereo.compute(right_gray, left_gray)

plt.figure(figsize=(7,7))
plt.imshow(disparity,'gray')
plt.title("Disparity Map")
plt.show()
```
