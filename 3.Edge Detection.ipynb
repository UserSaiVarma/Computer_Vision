```python
import cv2 as cv
import numpy as np
from PIL import Image
```

```python
image = cv.imread("images/resized_car.jpg")
Image.open("images/resized_car.jpg")
```

<font size=3 color='darkblue'>1. Sobel Filter

```python
def sobel_filter(image):
    "Applies Sobel Edge detection filter"
    image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    vertical_filter = np.array([[-1,0,1],
                                [-2,0,2],
                                [-1,0,1]])
    horizontal_filter = vertical_filter.transpose()
    vertical_filtered = cv.filter2D(image, -1, vertical_filter)
    horizontal_filtered = cv.filter2D(image, -1, horizontal_filter)
    abs_grad_x = cv.convertScaleAbs(vertical_filtered)
    abs_grad_y = cv.convertScaleAbs(horizontal_filtered)
    grad = cv.addWeighted(abs_grad_x, 0.5, abs_grad_y, 0.5, 0)
    return grad
```

```python
sobel_img = sobel_filter(image)
cv.imwrite("images/sobel_img.jpg",sobel_img)
Image.open("images/sobel_img.jpg")
```

<font size=3 color='darkblue'>2. Prewitt Filter

```python
def prewitt_filter(image):
    "Applies prewitt filter on the image"
    image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    vertical_filter = np.array([[-1,0,1],
                                [-1,0,1],
                                [-1,0,1]])
    horizontal_filter = vertical_filter.transpose()
    vertical_filtered = cv.filter2D(image, -1, vertical_filter)
    horizontal_filtered = cv.filter2D(image, -1, horizontal_filter)
    abs_grad_x = cv.convertScaleAbs(vertical_filtered)
    abs_grad_y = cv.convertScaleAbs(horizontal_filtered)
    grad = cv.addWeighted(abs_grad_x, 0.5, abs_grad_y, 0.5, 0)
    return grad
```

```python
prewitt_img = prewitt_filter(image)
cv.imwrite("images/prewitt_img.jpg",prewitt_img)
Image.open("images/prewitt_img.jpg")
```

```python
def laplacian_filter(image):
    "Applies prewitt filter on the image"
    image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    laplacian_filter = np.array([[1,1,1],
                                 [1,-8,1],
                                 [1,1,1]])
    filtered = cv.Laplacian(image, ksize=3, ddepth=cv.CV_16S)
    return cv.convertScaleAbs(filtered)
```

```python
laplace_img = laplacian_filter(image)
cv.imwrite("images/laplacian_img.jpg", laplace_img)
Image.open("images/laplacian_img.jpg")
```
